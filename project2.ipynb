```python
# ----------------- Student Management System (Deployed Version) ----------------- #

#  Imports
import csv
import os
import pandas as pd

# ----------------- 1. Setup CSV File ----------------- #
FILENAME = "students.csv"

# Create CSV file with headers if it does not exist
if not os.path.exists(FILENAME):
    with open(FILENAME, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["VID", "Name", "Citizen", "Course", "Status"])
        print("âœ” CSV file created with headers.")

# ----------------- 2. Define Functions ----------------- #

# Add a new student
def add_student(vid, name, citizen, course, status):
    with open(FILENAME, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([vid, name, citizen, course, status])
    print(f" Student '{name}' added successfully!")

# Read student by VID
def read_student(vid):
    df = pd.read_csv(FILENAME)
    student = df[df['VID'] == vid]
    if not student.empty:
        print(f"\n Student Record for VID {vid}:")
        print(student)
    else:
        print(f"No student found with VID {vid}")

# Update student record
def update_student(vid, name=None, citizen=None, course=None, status=None):
    df = pd.read_csv(FILENAME)
    if vid in df['VID'].values:
        if name: df.loc[df['VID'] == vid, 'Name'] = name
        if citizen: df.loc[df['VID'] == vid, 'Citizen'] = citizen
        if course: df.loc[df['VID'] == vid, 'Course'] = course
        if status: df.loc[df['VID'] == vid, 'Status'] = status
        df.to_csv(FILENAME, index=False)
        print(f" Student with VID {vid} updated successfully!")
    else:
        print(f" No student found with VID {vid}")

# Delete student record
def delete_student(vid):
    df = pd.read_csv(FILENAME)
    if vid in df['VID'].values:
        df = df[df['VID'] != vid]
        df.to_csv(FILENAME, index=False)
        print(f" Student with VID {vid} deleted successfully!")
    else:
        print(f"No student found with VID {vid}")

# Show all students
def show_all_students():
    df = pd.read_csv(FILENAME)
    if df.empty:
        print(" No student records found.")
    else:
        print("\n All Student Records:")
        print(df)

# ----------------- 3. Sample Data & Operations ----------------- #

# Add sample students
add_student(1, "Sirisha", "India", "CSE", "F1")
add_student(2, "Kavya", "India", "PHD", "F1")
add_student(3, "Renuka", "Germany", "Doctor", "F1")
add_student(4, "Megha", "Canada", "MCA", "F1")
add_student(5, "Ananya", "USA", "MBA", "F1")

# Display all students
show_all_students()

# Read a specific student
read_student(5)

# Update a student record
update_student(5, status="Approved")
read_student(5)  # Verify update

# Delete a student
delete_student(3)
show_all_students()  # Verify deletion

# ----------------- End of Deployed Project ----------------- #
```
